```python
class Tensorflow:
    """
    Create tensorflow functions
    """

    def __init__(self, shape, data):
        self.data = data
        self.shape = shape
        self.axon = data
#         self.shapeData = self.Tensor()



    def Tensor(self):
        """
        Create a tensor given 1-d array data and shape containing values for the tensor
        
        Shape
        Data

        """

        # Python program to multiply all values in the 
        # list using traversal 
  
        def multiplyList(myList) : 
      
    # Multiply elements one by one 
            result = 1
            for x in myList: 
                result = result * x  
            return result 


        lenShape = len(self.shape)
        lenData = len(self.data)

        minShapeSize = multiplyList(self.shape)	#Finding the minimum length data needs to be

        if minShapeSize  > lenData:					#Pad data with 0 if size of data is too small
            for i in range(minShapeSize - lenData):
                self.data.append(0)
            lenData = len(self.data)


        newData = self.data[:minShapeSize]		#Getting rid of extra data if any
        lenData = len(newData)					#Updating data length
        n = int(lenData / minShapeSize)
        
        self.shape.reverse()		#Working backward
        for j in self.shape:

            newData = [newData[i:i + j] for i in range(0, len(newData), j)]	#Split data up by # of element in the last array EX: 120 element w/ shape Length 40 has 3 element in last array


        returnData = newData.pop()
        return returnData


#Return the Tensor
    def printTensor(self):
        return self.shapeData
    
    def printAxon(self):
        return print("Axon: ", self.axon)
    def axonShape(self):
        return print("Axon Shape: ",self.axon.shape)

    def tensorMult(self, x):
        layers = len(self.shapeData)
        x_input = x 
        for numLayer in range(0, layers):
            axon = []
            weight = self.shapeData[numLayer]
            for w in weight:
                z = [i*j for i,j in zip(x_input, w)]
                axon.append(sum(z))
            print("Axon: ", axon)
            x_input = axon
        
        return axon


    def layer(self, weight):
        update_axon = []
        num=0
        for feature in self.axon:
            temp = []    #Used to store temporary feature prediction
            for w in weight:
                z = [i*j for i,j in zip(feature, w)]
                temp.append(sum(z))
            update_axon.append(temp)
            num+=1
#             print(num)
        self.axon = update_axon


#Resetting axon back to input data
    def resetAxon(self):
        self.axon = self.data


# 	def flatten(self, input):
#Change highest axon to 1 and rest to 0
    def softmax(self):
        for infer in self.axon:
            maxVal = max(infer)
            for i in range(len(infer)):
                if infer[i] == maxVal:
                    infer[i] = 1
                else:
                    infer[i] = 0
    
    
    def relu(self):
        for axons in self.axon:
            for i in range(len(axons)):
                if axons[i] < 0:
                    axons[i] = 0
                    
    def compare(self, y):
        
    
    
        error = 0
        for pred, label in zip(self.axon, y):
            print("pred: ",pred)
            print("label: ", label)
#             same = cmp(pred,label)
            if all(pred == label):
                error +=1
        
        return error




```
